```python
from datetime import datetime, timedelta 
windowSize = 3
b_max =[]
# ignore=[]

for i in range(4,len(result)):
    diff = result['cdt'].iloc[i] - result['cdt'].iloc[i-1]
    if (diff <= timedelta(hours=1)) :
        b_max.insert(i,max(result['normalize15'].iloc[i-windowSize:i+1]))
    else:
        b_max.insert(i,result['normalize15'].iloc[i])
        
print(len(b_max))
```

```python
windowSize = 3
b_min =[]

for i in range(4,len(result)):
    diff = result['cdt'].iloc[i] - result['cdt'].iloc[i-1]
    if (diff <= timedelta(hours=1)) :
        b_min.insert(i,min(result['normalize15'].iloc[i-windowSize:i+1]))
    else:
        b_min.insert(i,result['normalize15'].iloc[i])

print(len(b_min))
# print(b_min)
```

```python
final_result['max']=final_result['max'].astype(float)
final_result['min']=final_result['min'].astype(float)
# print(type(final_result['max']))
# print(type(final_result['min']))

final_result['Swing60'] = final_result['max'] - final_result['min']
final_result['Swing60KWH'] = final_result['Swing60'] * final_result['capacity'] / 4
final_result['Swing60KW'] = final_result['Swing60'] * final_result['capacity']

# print(final_result.head(20))
final_result.to_csv('/home/shraddha/Downloads/solar_dataset/meter_inverter_1102.csv')
print('^^')
```
